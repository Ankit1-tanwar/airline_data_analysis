```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv("/content/airlines.csv")
df.head(10)
```

```python
active_airline = df["Active"].value_counts().get("Y")
inactive_airline = df["Active"].value_counts().get("N")
print("Total no of active airline is ", active_airline)
print("Total no of inactive airline is  ", inactive_airline)
```

```python
colors = ["#dc3545", "#ffc107"]
plt.figure(figsize=(6,5))
plt.bar(["Active","Inactive"],[active_airline, inactive_airline], color = colors)
plt.xlabel('Airline Status')
plt.ylabel("No of airlines")
plt.title("comparsion of Active and Inactive airlines")
plt.show()
```

```python
# distribution of airline by Country
df_sorted = df.groupby('Country').size().to_frame(name='count').sort_values(by='count', ascending=False)
colors = ["lightcoral", "gold","mediumpurple","gray","darkseagreen","teal","crimson","steelblue","peru","lightgreen"]
top_n1 = df_sorted.index[:10]
top_n = df_sorted.loc[top_n1]
plt.figure(figsize=(12,6))
plt.bar(top_n.index, top_n["count"], color = colors)
plt.xlabel("country")
plt.ylabel("No of Airline")
plt.xticks(rotation=45, ha="right")
plt.title( f"{top_n.shape[0]}top counntry with most airline")
plt.show()
```

```python
df_sorted =df.groupby("Country").size().to_frame(name = "count").sort_values(by="count", ascending=True)
bottom_10_airline = df_sorted.tail(15).index
least_airline = df_sorted.loc[bottom_10_airline]
plt.plot(least_airline.index,least_airline["count"],marker="o",color="gold")
plt.xlabel("Country")
plt.ylabel("No of Airlines")
plt.xticks(rotation=45, ha="right")
plt.title("Top 10 Countries with least Airlines")
plt.show
```
